```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=15,10
```

```python

kk=np.linspace(-3,3,100)
kk=sorted(kk)


all_val=sum([np.exp(x) for x in kk])

def softmax(x,all_val=all_val):
    to_r=np.exp(x)/all_val
    return to_r    

def sigmoid(x):
    to_r=(1)/(1+np.exp(-x))
    return to_r

softmax_kk=[softmax(x) for x in kk]
sigmoid_kk=[sigmoid(x) for x in kk]

plt.scatter(kk,np.multiply(softmax_kk,10),label="softmax (scaled by lambda=10)")
plt.scatter(kk,sigmoid_kk,label="sigmoid")

plt.legend(loc="center left")
```
